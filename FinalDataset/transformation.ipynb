```python
from IPython.display import display
import pandas as pd

df_clientes = pd.read_csv("clientes.csv")
df_produtos = pd.read_csv("produtos.csv")
df_vendas = pd.read_csv("vendas.csv")

display(df_clientes)
```

```python
df_clientes['pais'] = df_clientes['pais'].replace({'PT': 'Portugal', 'Portugall': 'Portugal'})
df_clientes['data_registo'] = pd.to_datetime(df_clientes['data_registo'], format='mixed', errors='coerce')
df_clientes['data_registo'] = df_clientes['data_registo'].dt.strftime('%d-%m-%Y')
df_clientes['nome'] = df_clientes['nome'].str.replace('-', ' ')
df_clientes = df_clientes.reset_index(drop=True)
df_clientes['cliente_id'] = range(1, len(df_clientes) + 1)

# Verifica se o email é nulo ou inválido
def validar_email(email):
    if pd.isnull(email):
        return False
    return bool(pd.Series(email).str.contains(r'^[\w\.-]+@[\w\.-]+\.\w+$', regex=True).values[0])

# Aplica a correção
df_clientes['email'] = df_clientes['email'].apply(lambda x: x if validar_email(x) else 'sem email')

display(df_clientes)
```

```python
df_produtos = df_produtos.reset_index(drop=True)
df_produtos['produto_id'] = range(1, len(df_produtos) + 1)
df_produtos['stock'] = df_produtos['stock'].fillna(0).astype(int)
df_produtos['categoria'] = df_produtos['categoria'].fillna('Não Especificada')


display(df_produtos)
```

```python
df_vendas = df_vendas.reset_index(drop=True)
df_vendas['venda_id'] = range(1, len(df_vendas) + 1)
# Dicionário de conversão de palavras para números
extenso_para_numero = {
    'um': 1, 'dois': 2, 'três': 3, 'quatro': 4, 'cinco': 5,
    'seis': 6, 'sete': 7, 'oito': 8, 'nove': 9, 'dez': 10
}
# Corrigir a coluna 'quantidade'
def limpar_quantidade(valor):
    try:
        # Se for número, converte para inteiro e usa valor absoluto
        return abs(int(valor))
    except:
        # Se for texto por extenso, converte usando o dicionário
        valor_str = str(valor).strip().lower()
        return extenso_para_numero.get(valor_str, None)  # retorna None se não encontrar

# Aplicar a função
df_vendas['quantidade'] = df_vendas['quantidade'].apply(limpar_quantidade)
df_vendas['data_venda'] = pd.to_datetime(df_vendas['data_venda'], format='mixed', errors='coerce')
df_vendas['data_venda'] = df_vendas['data_venda'].dt.strftime('%d-%m-%Y')
df_vendas['metodo_pagamento'] = df_vendas['metodo_pagamento'].fillna('Não Pago')

display(df_vendas)
```
