```python
mapping = {
    '<1H OCEAN': 0,
    'INLAND': 1,
    'NEAR OCEAN': 2,
    'NEAR BAY': 3,
    'ISLAND': 4
}

data["ocean_proximity_num"] = data["ocean_proximity"].map(mapping)
```

```python
data[["ocean_proximity", "ocean_proximity_num"]].head()
```

```python
data.drop('ocean_proximity', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value', 'ocean_proximity'], axis=1) #attributes
y = data['median_house_value']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred =  lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("Accuracy:", r2_score(y_test, y_pred))
```

```python
new_data = [[-120, 34, 20, 100, 20, 1000, 150, 4.5, 3]]
prediction = lr.predict(new_data)
print("Predicted median house value:", prediction)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
from sklearn.metrics import r2_score

print("Accuracy:", r2_score(y_test, y_pred))
```

```python
new_data = [[-120, 34, 20, 100, 20, 1000, 150, 4.5, 3]]
prediction = rf.predict(new_data)
print("Predicted median house value:", prediction)
```
